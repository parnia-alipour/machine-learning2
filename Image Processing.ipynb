```python
import numpy as np
import pandas as pd
import matplotlib_inline
```

```python
x=pd.read_csv('training.csv')
x[:5]
```

```python
from  sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['class']=le.fit_transform(x['class'])
x['class']
```

```python
feature=x[['GLCM_pan','Mean_Green','Mean_Red','Mean_NIR','SD_pan']]
feature[:10]
```

```python
y=x['class'].values
y
```

```python
x=np.asanyarray(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(feature,y,test_size=0.2,random_state=42)

print('set training data:',x_train.shape,y_train.shape)
print('set testing data:',x_test.shape,y_test.shape)
```

```python
from sklearn.preprocessing import PowerTransformer
scaler=PowerTransformer()
x_train1=scaler.fit_transform(x_train)
x_test1=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression().fit(x_train1,y_train)
regressor
```

```python
test_image=pd.read_csv('testing.csv')
test_image[:5]
```

```python
y_pred=regressor.predict(x_test1)
y_pred
```

```python
from sklearn.metrics import r2_score ,accuracy_score,precision_score,recall_score


print('accuracy:',accuracy_score(y_test,y_pred))
print('precision_score:',precision_score(y_test,y_pred))
print('recall_score:',recall_score(y_test,y_pred))
```
